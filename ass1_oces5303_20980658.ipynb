# 1. Introduction & Motivation

Initially, this project aimed to predict the El Niño-Southern Oscillation (ENSO) using time-series data. However, a critical machine learning constraint was identified: the effective sample size. Because ENSO operates on a 2-7 year cycle, the ~20-year span of modern continuous data contains merely a handful of independent events. This severely limits statistical validity and induces temporal data leakage. To ensure robust model evaluation while fully leveraging high-dimensional oceanographic data, this project pivoted to a spatial classification task: Global Ocean Basin Identification.





Traditionally, ocean water masses have been defined using subjective regional bounding boxes and manual Temperature-Salinity (T-S) curve analyses (Emery, 2001). However, the deployment of the global Argo array (Roemmich et al., 2009) has provided an unprecedented volume of high-resolution vertical profiles, necessitating data-driven approaches.

Recent literature has demonstrated the power of machine learning in physical oceanography. Maze et al. (2017) pioneered the use of unsupervised learning to extract coherent thermodynamic patterns directly from Argo profiles, proving that water masses have intrinsic mathematical structures. Similarly, Sonnewald et al. (2019) utilized machine learning to objectively define robust physical and dynamical boundaries in the ocean, moving away from arbitrary geographic borders.

Building upon this paradigm, this project employs a "Proxy Task": We deliberately withhold spatial metadata (Latitude/Longitude) and task a Random Forest classifier to predict the source ocean basin (Atlantic, Pacific, Indian) entirely from a 1D vertical physical profile. If the model succeeds, it objectively proves that distinct basins possess unique, quantifiable "Water Mass Fingerprints" that machine learning can detect.

While modern floats provide high-resolution data (e.g., 200+ vertical layers), maintaining such sensor arrays is economically expensive. We introduce a novel constraint to this project: Sparse Sampling. We drastically downsample the Argo profiles to a coarse 100-meter interval from the surface down to 1500m (reducing features from over 200 to just 48).

Furthermore, we conduct a physical ablation study, by isolating Thermodynamics (Temperature), Salinity proxies (Density/SIG0), and Dynamics (Stratification/BRV2)—to investigate the fundamental physical drivers of the model.

This assignment is aims to:

1. Train a Random Forest model on sparse (100m interval) vertical profiles to classify ocean basins.

2. Utilize Grid Search Cross-Validation to optimize hyperparameters while maintaining high computational efficiency.

3. Perform a spatial diagnostic of misclassifications to trace real-world physical mixing zones (e.g., the Agulhas Leakage and Indonesian Throughflow).

# 2. Data Acquisition & Preprocessing

2.1 Data Source & Variables

The hydrographic profiles are sourced from the continuous, homogeneous global Argo .zarr dataset. Before the 21st century, physical oceanography relied heavily on sparse, ship-based measurements, which introduced severe spatial and seasonal biases. This paradigm was revolutionized by the Argo Program (Roemmich et al., 2009), an international collaborative observation network.

<img src="https://argo.ucsd.edu/wp-content/uploads/sites/361/2020/06/float_cycle_1-768x424.png" width="600" alt='Argo'>

The Argo array consists of nearly 4,000 free-drifting autonomous profiling floats distributed across the global ocean. These instruments continuously monitor the upper 2,000 meters of the water column. By parking at a drift depth (typically 1,000m), diving to 2,000m, and subsequently ascending to the surface while recording continuous thermodynamic profiles, Argo floats transmit real-time data via satellite.  This provides an unprecedented, homogeneous, high-resolution 3D dataset of the global ocean's state, serving as the foundational data source for this machine learning study.


The dataset utilized in this project (GLOB_HOMOGENEOUS_variables.zarr) extracts three fundamental physical parameters from the Argo profiles. Rather than feeding raw, uninterpreted data into the Random Forest, we selected these variables because each serves as a distinct physical proxy for defining ocean basin characteristics:
1. TEMP (In-situ Temperature, °C): This represents the sheer thermodynamic state of the water column. While upper-ocean temperature gradients are heavily influenced by atmospheric forcing and latitudinal solar radiation, deep-water temperatures act as conservative tracers that track the origins of global ocean circulation.
2. SIG0 (Potential Density Anomaly, $\sigma_0$): Density in the ocean is a non-linear function of temperature, salinity, and pressure. SIG0 represents the density of a seawater parcel if it were brought adiabatically (without heat exchange) to the sea surface (pressure = 0 dbar), minus a reference value of $1000 \, \text{kg/m}^3$. Because it implicitly captures the salinity footprint, SIG0 is the primary diagnostic variable used by physical oceanographers to identify distinct water masses (e.g., the dense North Atlantic Deep Water).
3. BRV2 (Brunt-Väisälä Frequency Squared, $N^2$): This is a core metric of ocean dynamics. It quantifies the static stability, or stratification, of the water column. A high $N^2$ value indicates strong stratification (such as within the thermocline/pycnocline), which severely suppresses vertical fluid mixing. Conversely, values near zero indicate perfectly mixed layers. By including BRV2, we provide the machine learning model with the dynamic, vertical "texture" of the water column.

2.2 Coordinate Boundary Transformation
A classic geospatial data artifact occurs at the International Date Line. The original dataset defines longitude on a [-180°, 180°] grid. This artificial boundary splits the Pacific Ocean into two disconnected numerical domains (positive and negative), which would catastrophically confuse the spatial labeling function. To resolve this, a modulo transformation (Longitude + 360) % 360 was applied, mapping the global ocean onto a continuous [0°, 360°] grid.

2.3 Target Generation & Feature Flattening

Target Variable (y): The continuous [0°, 360°] longitudes were mapped to three discrete categorical labels based on major basin boundaries: Atlantic, Indian, and Pacific.

Feature Matrix (X): The 3D spatio-temporal dataset was flattened into a 2D Machine Learning matrix. Each profile (row) represents a distinct water column, and each column represents a physical variable at a specific depth layer (e.g., TEMP_100m, SIG0_1500m).

2.4 Sparse Sampling & Data Leakage Prevention
To fulfill the "Data Efficiency" objective, we applied a rigid downsampling mask. Only depth layers from 0m to 1500m at strict 100-meter intervals were retained. This aggressive dimensionality reduction shrank the feature space from over 200 variables down to exactly 48 sparse physical features.
Critically, to prevent spatial data leakage, geographic coordinates (LATITUDE, LONGITUDE) were strictly dropped from the training features, ensuring the model learns pure physics, not geographic memorization.

```python
# Load Dataset
ds = xr.open_zarr("../../GLOB_HOMOGENEOUS_variables.zarr/", consolidated=False)
ds_global = ds.where(((ds['LATITUDE'] >= -50) & (ds['LATITUDE'] <= 50)).compute(), drop=True)

df_profile = ds_global[['TEMP', 'SIG0', 'BRV2']].to_dataframe().reset_index()
df_profile = df_profile.dropna(subset=['TIME', 'LATITUDE', 'LONGITUDE'])

df_profile['LONGITUDE'] = (df_profile['LONGITUDE'] + 360) % 360

# Define Ocean Basin
def get_ocean_basin(lon):
    if 20 <= lon < 120: return 'Indian'
    elif 120 <= lon < 290: return 'Pacific'
    else: return 'Atlantic'

df_profile['BASIN'] = df_profile['LONGITUDE'].apply(get_ocean_basin)

# Pivot to wide format
X_raw = df_profile.pivot_table(
    index=['TIME', 'LATITUDE', 'LONGITUDE', 'BASIN'], 
    columns='DEPTH', values=['TEMP', 'SIG0', 'BRV2']
).reset_index()

X_raw.columns = [f"{c[0]}_{c[1]}m" if c[1] != '' else c[0] for c in X_raw.columns]
X_ml = X_raw.dropna()

sparse_cols = ['TIME', 'LATITUDE', 'LONGITUDE', 'BASIN']
features = []
for col in X_raw.columns:
    if col not in sparse_cols and '_' in col:
        depth_val = abs(float(col.split('_')[1].replace('m', '')))
        if depth_val <= 1500 and depth_val % 100 == 0:
            sparse_cols.append(col)
            features.append(col)

X_ml = X_raw[sparse_cols].dropna()
print(f"Sparse sampling complete! Features reduced to {len(features)}.")
print(f"Total profiles: {len(X_ml)}")
```

# 3. Methodology

3.1 Exploratory Data Analysis : Linear vs. Non-linear Manifolds
Before training the supervised model, we visually inspect the 48-dimensional sparse feature space. We employ Principal Component Analysis (PCA) to test for linear separability, and t-Distributed Stochastic Neighbor Embedding (t-SNE) to explore non-linear manifolds. This step verifies whether the downsampled profiles still retain distinct clustering structures corresponding to different ocean basins.

3.2 Model Selection: Random Forest Classifier
A Random Forest Classifier (Breiman, 2001) is selected as the core algorithm for this study. Unlike deep neural networks, tree-based ensemble methods are highly robust to feature scaling, handle non-linear oceanic relationships effectively, and most importantly, provide interpretable Feature Importance. This interpretability is vital for extracting physical oceanographic insights (i.e., identifying which depth and variable dominates the water mass fingerprint).

3.3 Training Pipeline & Hyperparameter Tuning
The dataset is split into training (80%) and testing (20%) sets using a stratified split. Because this is a spatial classification task rather than a time-series forecast, random splitting is physically valid and devoid of the temporal leakage issues associated with ENSO predictions.

To optimize the model while maintaining computational efficiency on the sparse dataset, we implement a Grid Search with 3-Fold Cross-Validation (GridSearchCV). We constrain the search space to essential hyperparameters (n_estimators, max_depth) to prevent memory overflow while securing the optimal estimator. All data preprocessing, model training, and Grid Search cross-validation pipelines were implemented using the Scikit-learn Python library (Pedregosa et al., 2011).

3.4 Physical Ablation Study Design
To systematically deconstruct the physical drivers of the model, we design an ablation study consisting of four separate Random Forest models trained on subsets of the 48 sparse features:

1. Thermodynamics Only: Trained exclusively on TEMP features.

2. Density/Salinity Only: Trained exclusively on SIG0 features.

3. Dynamics Only: Trained exclusively on BRV2 features.

4. The Full Fingerprint: Trained on all 48 combined features.

3.5 Evaluation Strategy: Spatial Error Diagnostics
Beyond standard machine learning metrics (Accuracy and Confusion Matrix), this study employs a domain-specific evaluation method: Spatial Diagnostic Mapping. By mapping the misclassified test samples back onto global coordinates, we can determine whether the model's errors are arbitrary statistical noise, or whether they reflect true, physical water mass mixing zones (e.g., inter-basin exchange pathways).

4.2 Model Performance & Hyperparameter Optimization

To establish a performance floor, a baseline Logistic Regression model was evaluated using 5-Fold Cross-Validation, achieving a robust mean accuracy of 93.74% (±0.17%). This indicates that the sparse physical features are fundamentally highly predictive.

We then trained a Random Forest classifier. Utilizing GridSearchCV on the sparse dataset allowed for extremely fast optimization (fitting 6 candidates in seconds). The optimal hyperparameters were identified as n_estimators=100 and max_depth=15. This optimized model achieved an outstanding test accuracy of 98.53%.
The accompanying Confusion Matrix demonstrates near-perfect classification across all classes. The heavily dominated diagonal confirms minimal cross-basin leakage, with the model confidently identifying the massive Pacific dataset (over 11,000 correct samples) with negligible errors.

```python
# 1. Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2, random_state=42, stratify=y_labels)

# 2. Feature Standardization
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. Build Baseline Model (Logistic Regression)
lr = LogisticRegression(max_iter=1000, random_state=42)

# 5-Fold Cross Validation
cv_scores = cross_val_score(lr, X_train_scaled, y_train, cv=5, scoring='accuracy')

print("=== Baseline Model (Logistic Regression) ===")
print(f"5-Fold Cross-Validation Scores: {cv_scores}")
print(f"Mean CV Accuracy: {cv_scores.mean()*100:.2f}% (±{cv_scores.std()*100:.2f}%)")

lr.fit(X_train_scaled, y_train)
print(f"Test Set Accuracy: {accuracy_score(y_test, lr.predict(X_test_scaled))*100:.2f}%")
```

```python
X_features = X_ml[features]
y_labels = X_ml['BASIN']

# Split 80% train dataset (tr) and 20% test dataset (te)
X_tr, X_te, y_tr, y_te = train_test_split(
    X_features, y_labels, test_size=0.2, random_state=42, stratify=y_labels
)

param_grid = {
    'n_estimators': [50, 100],  
    'max_depth': [15]           
}

# 2 parameter groups * 3 croos validation
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy', n_jobs=1, verbose=1)

grid_search.fit(X_tr, y_tr)

# Select the best model and evaluate on test set
best_rf = grid_search.best_estimator_
y_pred_rf = best_rf.predict(X_te)

print(f" Best Hyperparameters: {grid_search.best_params_}")
print(f" Optimized Test Accuracy: {accuracy_score(y_te, y_pred_rf)*100:.2f}%\n")
```

4.3 hysical Interpretability: The Deep Density Signature and Exchange Chokepoints

Machine learning in oceanography must be physically interpretable. Rather than acting as a statistical "black box," the Random Forest autonomously rediscovered fundamental principles of descriptive physical oceanography. The diagnostics of feature importance and spatial misclassifications reveal exactly how the model physically distinguishes the global basins.

4.3.1 The Memory of Deep Water Masses (Feature Importance)

Remarkably, the feature importance depth profile demonstrates that the model assigns the highest Gini importance to potential density ($\sigma_0$) and temperature at deep layers (1000m - 1400m), while largely ignoring surface metrics and stratification ($N^2$). This is dynamically brilliant. The upper ocean (0-200m) is heavily contaminated by local atmospheric forcing (solar radiation, wind stress, and freshwater flux), making surface signatures noisy and ambiguous. Stratification ($N^2$) is also ignored because static stability—characterized by a strong main pycnocline and well-mixed deep layers—is a universal feature across all mid-to-low latitude basins, thus lacking tracer utility.

Instead, the model's heavy reliance on deep ocean density accurately reflects the classical oceanographic method of Water Mass Analysis. The Atlantic basin is uniquely characterized by the formation of dense North Atlantic Deep Water (NADW), making it significantly saltier and denser at depth compared to the Pacific and Indian Oceans (Broecker, 1991; Talley et al., 2011). By locking onto this deep thermohaline gradient as the ultimate basin discriminator, the model proves it learned genuine, long-term physical "memory" rather than superficial climate noise.

4.3.2 "Errors" as Physical Oceanographic Realities (Spatial Diagnostic)

Furthermore, mapping the spatial distribution of the model's "misclassifications" proves that these are not statistical failures, but successful identifications of intense interocean water mass exchange. Geographic boundaries (like longitudes) are human constructs, whereas the model predicts based on true thermodynamic signatures.

The four most prominent clusters of classification errors align perfectly with global ocean chokepoints:

1. The Agulhas Leakage: The dense cluster of errors south of Africa (near 30°E) represents warm, salty Indian Ocean water leaking into the South Atlantic via Agulhas Rings (Beal et al., 2011). The model accurately detected the Indian Ocean thermohaline signature here, causing a geographic mismatch with the Atlantic label.

2. The Indonesian Throughflow (ITF): Errors clustered in the Indonesian seas (near 120°E) represent the continuous, massive advection of Pacific upper-ocean waters into the Indian Ocean (Gordon, 2005). The model's "confusion" here perfectly mirrors this physical flow.

3. The ACC Mixing Bowl: Crucially, the continuous band of misclassifications along the southern edge of the domain (~45°S - 50°S) directly traces the path of the Antarctic Circumpolar Current. As the only ocean current that flows completely around the globe without continental barriers, the ACC acts as the primary mixing bowl for the world's oceans (Rintoul, 2018). It constantly blends Atlantic, Indian, and Pacific water masses through intense mesoscale eddies and frontal dynamics. The model's inability to strictly categorize profiles in this region correctly reflects the physical reality: in the ACC, water parcels belong to a global, interconnected system rather than a single isolated basin.

4. Western Boundary Currents: A localized cluster of errors off the east coast of North America suggests the model occasionally confuses the intense, eddy-rich Gulf Stream with its dynamic Pacific twin, the Kuroshio current, due to their remarkably similar kinetic and thermodynamic structures.

Ultimately, the model did not just memorize a map; it mapped the physical reality of the global thermohaline circulation.

# 5. Conclusions & Acknowledgements

5.1 Synthesizing the Sparse Fingerprint Hypothesis
This study successfully demonstrated that machine learning algorithms can objectively classify global ocean basins based solely on vertical hydrographic profiles, entirely independent of geographic coordinates. Crucially, by restricting the Random Forest model to a highly sparse observational grid (only 48 features sampled at 100-meter intervals), we proved that the macroscopic thermohaline backbone of the ocean is extraordinarily robust. The optimized model achieved an outstanding accuracy of 98.53%, validating the hypothesis that distinct ocean basins possess intrinsic, mathematically separable "fingerprints" that do not require ultra-high-resolution arrays to be detected (Maze et al., 2017).

5.2 Opening the Black Box: Physics over Statistics
Moving beyond "black-box" predictions, our feature importance analysis provided profound oceanographic interpretability. The ablation study and Gini impurity metrics revealed that the model autonomously identified deep-water density (SIG0 at >1000m) as the dominant discriminator. This aligns perfectly with classical physical oceanography, successfully capturing the dense signature of North Atlantic Deep Water (NADW) (Broecker, 1991; Talley et al., 2011). It confirms that the algorithm anchored its decisions on fundamental ocean dynamics rather than superficial, latitudinally symmetric surface temperatures.

5.3 The Paradigm of Errors as Physical Mixing
Perhaps the most compelling finding emerged from the spatial diagnostic of the model's misclassifications. Rather than exhibiting random statistical noise, the errors were heavily clustered at known inter-basin exchange choke points—specifically, the Indonesian Throughflow and the Agulhas Leakage (Sprintall et al., 2014; Beal et al., 2011). The model’s mathematical "confusion" elegantly mirrors the real-world physical mixing of water masses. This reinforces the paradigm that machine learning can be utilized not just to classify, but to dynamically map fluid boundaries and global ocean conveyor pathways where traditional geographic borders fail (Sonnewald et al., 2019).
